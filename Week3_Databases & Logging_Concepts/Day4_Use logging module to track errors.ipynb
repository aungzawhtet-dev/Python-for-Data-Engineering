```python
import logging
import sqlite3

# Configure logging
logging.basicConfig(
    filename="app.log",
    level=logging.INFO, # Set logging level to INFO. This sets the minimum severity of messages to log.
    format="%(asctime)s - %(levelname)s - %(message)s" # Log format with timestamp,Log level, and message
)

try:
    conn = sqlite3.connect("practice.db") # Connect to SQLite database stored in practice.db
    cur = conn.cursor() # Create a cursor object to execute SQL commands
    cur.execute("SELECT * FROM non_existing_table") # Attempt to execute a query on a non-existing table
except sqlite3.Error as e:
    logging.error("Database error occurred: %s", e) # %s is a placeholder for the error message
except Exception as e:
    logging.exception("Unexpected error: %s", e)
else:
    logging.info("Query executed successfully")
finally:
    conn.close()
    
print("Script completed. Check app.log for details.")
```
