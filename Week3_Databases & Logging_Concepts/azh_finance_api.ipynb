```python
finance_api_response = [
    {
        "symbol": "AAPL",
        "price": 172.3,
        "open": 170.0,
        "high": 173.5,
        "low": 169.8,
        "volume": 50000000,
        "timestamp": 1694937600
    },
    {
        "symbol": "TSLA",
        "price": 250.0,
        "open": 248.5,
        "high": None,
        "low": 245.0,
        "volume": None,
        "timestamp": None
    }
]


# convert the API response to a DataFrame
# before inserting into the dataframe, flatten the JSON data

records = []
for item in finance_api_response:
    record = {
        "symbol": item["symbol"],
        "price": item["price"],
        "open": item["open"],
        "high": item["high"],
        "low": item["low"],
        "volume": item["volume"],
        "timestamp": item["timestamp"]
    }
    records.append(record)

df = pd.DataFrame(records)
df
```
