```python
import requests # for making API requests
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime
import json
```

```python
youtube_api_response = {
    "kind": "youtube#searchListResponse",
    "items": [
        {
            "id": {"videoId": "abc123"},
            "snippet": {
                "publishedAt": "2023-09-17T08:00:00Z",
                "channelId": "UC123",
                "title": "Data Engineering Tutorial",
                "description": "Learn ETL pipelines step by step",
                "channelTitle": "Data Science Academy"
            }
        },
        {
            "id": {"videoId": "def456"},
            "snippet": {
                "publishedAt": None,
                "channelId": "UC456",
                "title": "Python Tips",
                "description": None,
                "channelTitle": "Python Hub"
            }
        }
    ]
}


# Transform: Convert JSON to Pandas DataFrame
flatten_records = []
for item in youtube_api_response.get("items", []):
    record = {
        
        "video_id": item.get("id", {}).get("videoId"),
        "published_at":item.get("snippet", {}).get("publishedAt"),
        "channel_id": item.get("snippet", {}).get("channelId"),
        "title": item.get("snippet", {}).get("title"),
        "description": item.get("snippet", {}).get("description"),
        "channel_title": item.get("snippet", {}).get("channelTitle"),
         }
    flatten_records.append(record)
    print(flatten_records) # debugging purposes
    
# Print result
# for r in flatten_records :
#     print(r) # for debugging 
```

```python
# svare the list result,flatten_records  to a dataframe
youtube_df = pd.DataFrame(flatten_records)
youtube_df
```
