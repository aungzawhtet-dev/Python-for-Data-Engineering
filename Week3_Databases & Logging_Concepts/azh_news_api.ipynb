```python
import requests # for making API requests
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime
import json
```

```python
# 1. Extract: Simulated API response and store in variable
# Sample JSON data simulating API response

news_api_response = {
    "status": "ok",
    "totalResults": 2,
    "articles": [  # news API response is dictionary --> so we article
        {
            "source": {"id": "bbc-news", "name": "BBC News"},
            "author": "BBC Reporter",
            "title": "Global markets rally",
            "description": "Tech giants lead stock markets.",
            "url": "https://www.bbc.com/news/markets",
            "publishedAt": "2023-09-17T12:34:56Z",
        },
        {
            "source": {"id": None, "name": "Reuters"},
            "author": None,
            "title": "Oil prices climb",
            "description": "Crude oil prices rise due to supply.",
            "url": "https://www.reuters.com/business/energy",
            "publishedAt": None
        }
    ]
}



# 2. Transform: Convert JSON to Pandas DataFrame
flattened_records  = []
for article in news_api_response.get("articles", []):
    record = {
         "status": news_api_response.get("status"),
        "totalResults": news_api_response.get("totalResults"),
        "source_id": article.get("source",{}).get("id"),        
        "source_name":article.get("source",{}).get("name"),
        "author":article.get("author"),
        "title":article.get("title"),
        "url":article.get("url"),
        "publishedAt":article.get("publishedAt"),
           }
    flattened_records .append(record)
    
# Print result
for r in flattened_records :
    print(r) # for debugging 
```

```python
# save into Json file
with open ("flattened_records.txt","w",encoding="utf-8") as f:
    for record in flattened_records:
        f.write(json.dumps(record,ensure_ascii= False) + "\n") # If article text has non-English characters, theyâ€™ll be saved as-is instead of \u0e02\u0e48....
        
print("successfully generated flattened_records text file")
```
