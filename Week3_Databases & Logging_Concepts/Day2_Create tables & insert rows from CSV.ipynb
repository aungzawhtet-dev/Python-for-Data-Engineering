```python
import csv
import sqlite3

# SQLite example
conn = sqlite3.connect("practice.db")
cur = conn.cursor()

# Create table
cur.execute("""
CREATE TABLE IF NOT EXISTS employees (
    emp_id INTEGER PRIMARY KEY,
    emp_name TEXT,
    dept TEXT,
    salary integer(10)
)
""")

# Insert from CSV
with open("employees.csv", "r") as f:
    reader = csv.DictReader(f) # DictReader will read the CSV into a dictionary format
    for row in reader:
        cur.execute("""
        INSERT OR REPLACE INTO employees (emp_id, emp_name, dept, salary)
        VALUES (:emp_id, :emp_name, :dept, :salary) --- the placeholders are named to match the CSV headers
        """, row) # the row is passed as a dictionary

conn.commit() # Commit the transaction
conn.close() # Close the connection
print("CSV data inserted into employees table")
```
