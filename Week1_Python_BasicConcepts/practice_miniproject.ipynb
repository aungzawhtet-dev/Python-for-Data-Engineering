```python
import os
import csv


def combine_csvs(input_folder, output_file):
    all_rows = []   # list to hold all rows from all CSVs
    seen = set()    # set to track seen rows for deduplication
    header = None   # to store header once

    try:
        # 2. loop through the files in the folder
        for file in os.listdir(input_folder):
            if file.endswith('.csv'):  # fixed typo
                file_path = os.path.join(input_folder, file)
                print(f"Reading file: {file_path}")
                
                # 3. open each csv file
                with open(file_path, "r", newline="") as csvfile:
                    reader = csv.reader(csvfile)
                    file_header = next(reader)  # read the first line (header)

                    if header is None:
                        header = file_header

                    # 4. read rows and remove duplicates
                    for row in reader:
                        row_tuple = tuple(row)
                        if row_tuple not in seen:
                            seen.add(row_tuple)
                            all_rows.append(row)

        # 5.Write combined rows to output file (after reading all files)
        with open(output_file, "w", newline="") as csvfile:
            writer = csv.writer(csvfile)
            if header:
                writer.writerow(header)
            writer.writerows(all_rows)

        print(f"Combined CSV saved to {output_file}")

    except Exception as e:
        print(f"Error: {str(e)}")


# Example usage
if __name__ == "__main__":
    combine_csvs("data_folder", "output_current.csv")
```
